## <a id='3'>3. Theoretical framework</a>

**Supervised learning**

Supervised learning, as the name indicates, has the presence of a supervisor as a teacher. Basically supervised learning is when we teach or train the machine using data that are well labeled. This means some data is already tagged with the correct answer. After that, the machine is provided with a new set of examples(data) so that the supervised learning algorithm analyses the training data(set of training examples) and produces a correct outcome from labeled data.

**Classification**

A classification problem is when the output variable is a category, such as Red or blue, disease or no disease



*Definitions extracted from [IRIS Classification a supervised approach](https://medium.com/@adityakataria36/iris-classification-an-supervised-approach-7b1eaad1711e) in Medium.com*

**Supervised classifications algorithms**

-----------------------------------------------------------------
**Logistic Regression**

Is a Machine Learning algorithm that is used for classification problems and is based on the concept of probability. It is used when the dependent variable (target) is categorical. It is widely used when the classification problem at hand is binary; true or false, yes or no, etc. Logistics regression uses the sigmoid function to return the probability of a label.



-----------------------------------------------------------------

**Decision tree**

Decision Trees are a non-parametric supervised learning method used for classification and regression. The overall end goal is to build a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.



-----------------------------------------------------------------

**Random forest**

Random Forests is a Supervised Learning algorithm and is made up of many decision trees. A good way to remember it is a multitude of trees building up a forest.
The random forest algorithm produces its outcomes based on the predictions generated by the Decision Trees. This prediction is done by taking the average or mean of the output from the various Decision Trees.



-----------------------------------------------------------------

**KNN (K-Nearest Neighbor)**

Is a supervised machine learning algorithm that can be used to solve both classification and regression problems. The KNN algorithm assumes that similar things exist in close proximity.
KNN uses the idea of similarity, or other words distance, proximity, or closeness. It uses a mathematical approach to calculate the distance between points on a graph. Doing so, it then labels the unobserved data based on the nearest labeled observed data points.

)

-----------------------------------------------------------------

**Support Vector Machine (SVM)**

Is a supervised learning model using a linear model that can be used for classification and regression problems.
The concept of the Support Vector Machine algorithm creates a line or a hyperplane which separates the data into classes. It uses the data points that are closer to the hyperplane and influence the position and orientation of the hyperplane; which in turn maximizes the margin of the classifier.

### Multivariate analysis

We need to validate if a variable has influence in the risk or not, this could be done viewing the proportion of people how are in risk or not for variable.

```python
# Create a function to graph the variables compared vs the risk 

def StackedBarchart(variable):
    cross_tab_prop = pd.crosstab(index=df[variable],
                                 columns=df['risk_flag'],
                                 normalize="index")

    cross_tab = pd.crosstab(index=df[variable],
                            columns=df['risk_flag'])

    cross_tab_prop.plot(kind='bar', 
                        stacked=True, 
                        colormap='tab10', 
                        figsize=(10, 6))

    plt.legend(loc="upper left", ncol=1)
    plt.xlabel(variable)
    plt.ylabel("Proportion")

    for n, x in enumerate([*cross_tab.index.values]):
        for (proportion, y_loc) in zip(cross_tab_prop.loc[x],
                                       cross_tab_prop.loc[x].cumsum()):

            plt.text(x=n - 0.1,
                     y=(y_loc - proportion) + (proportion / 2),
                     s=f'{np.round(proportion * 100, 1)}%', 
                     color="black",
                     fontsize=12,
                     fontweight="bold")
    plt.title('Stacked barchart for ' + variable + ' and risk_flag')
    return plt.show()
```

```python
# Stacked bar for the marital status
StackedBarchart('married_single')
```

The difference in proportion between at-risk individuals who are married and those who are not is close to 2.3%, so marital status is not expected to influence loan risk. So we have to run a chi-square test to validate if exist a independency between variables or not.

```python
# defining a function to calculate the p-value of the chi square test
def chisquare(variable):
    # defining the table
    data = pd.crosstab(df[variable], df["risk_flag"])
    stat, p, dof, expected = chi2_contingency(data)

    # interpret p-value
    alpha = 0.05
    print("p value is " + str(p))
    if p <= alpha:
        print(f'Dependent (reject H0) --> The variables "risk flag" and {variable} have a significant relation')
    else:
        print(f'Independent (H0 holds true) --> The variables "risk flag" and {variable} dont have a significant relation')
```

```python
# Chi square test for risk and married_single
chisquare('married_single')
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[['income','age','experience','current_job_yrs', 'current_house_yrs']] = scaler.fit_transform(X_train[['income','age', 'experience', 'current_job_yrs', 'current_house_yrs']])
X_test_scaled[['income','age','experience','current_job_yrs', 'current_house_yrs']] = scaler.transform(X_test[['income','age','experience','current_job_yrs', 'current_house_yrs']])

C_parameters = [0.001, 0.01, 0.1, 1, 10, 100]

train = []
test = []

for C_parameter in C_parameters:
    LR = LogisticRegression(C = C_parameter, max_iter = 10000).fit(X_train_scaled, y_train)

    print(f'Results with c parameter: {C_parameter}')
    print('Accuracy of Logistic regression classifier on training set: {:.2f}'
         .format(LR.score(X_train_scaled, y_train)))
    print('Accuracy of Logistic regression classifier on test set: {:.2f}'
         .format(LR.score(X_test_scaled, y_test)))
    print('\n')
    
    
    train.append(LR.score(X_train_scaled, y_train))
    test.append(LR.score(X_test_scaled, y_test))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpatches

train_label = mpatches.Patch(color='orange', label='Train accuracy')
test_label = mpatches.Patch(color='blue', label='Test accuracy')

plt.plot(C_parameters, test)
plt.plot(C_parameters, train)
plt.suptitle('Logistic regression')
plt.title('Test and train accuracy vs c parameter')
plt.legend(loc="upper left", ncol=1, handles=[train_label, test_label])
plt.xlabel('c')
plt.ylabel("Accuracy")
plt.show()
```

### Model selection

The model with the highest accuracy for the training and test data set is the Decision Tree algorithm and a maximum depth of 50, but it should be noted that this model is complex compared to simpler models that offer close accuracy, such as a Logistic Regression with C = 1. 

Because accuracy is a global metric that does not show the actual behavior of the model with test data, it is necessary to use other metrics such as:

- Accuracy = TP + TN / (TP + TN + FP + FN)
- Precision = TP / (TP + FP)
- Recall = TP / (TP + FN)  Also known as sensitivity, or True Positive Rate
- F1 = 2 * Precision * Recall / (Precision + Recall) 

With this metric we can observe the difference between the models and select the one that for our purpose is better, in a scenario like this we will assume that the best model is the one with the highest recall.

Recall or True Positive Rate (TPR) is the fraction of all positive instances that the classifier correctly identifies as positive.
